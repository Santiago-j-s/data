```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_json('tweets.json')
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

tweets = pd.Series(1, index=df.created_at)
tweets_per_hour = tweets.groupby(tweets.index.hour).count()
tweets_per_weekday = tweets.groupby(tweets.index.weekday_name).count().reindex(days)
tweets_per_day = tweets.groupby(tweets.index.day).count()
```

```python
sns.set(style="white")

fig, (ax1, ax2, ax3) = plt.subplots(figsize=(15, 12), nrows=3)
fig.suptitle('Número de tuits', fontsize=16)

sns.barplot(
    ax = ax1,
    x = tweets_per_hour.index,
    y = tweets_per_hour.values,
    color='b'
)
ax1.set(
    title='por hora',
    xlabel=''
)

dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
sns.barplot(
    ax = ax2,
    x = dias,
    y = tweets_per_weekday.values,
    color='b'
)
ax2.set(
    title='por día de la semana'
)

sns.barplot(
    ax = ax3,
    x = tweets_per_day.index,
    y = tweets_per_day.values,
    color='b'
)
ax3.set(
    title='por día del mes',
    xlabel=''
)

fig.subplots_adjust(hspace=0.5)
plt.show()
```
