```python
db_path = 'COVID-19/csse_covid_19_data/csse_covid_19_time_series'

indexes = [
    'Province/State',
    'Country/Region',
    'Lat',
    'Long',
]

df_confirmed = pd.read_csv(os.path.join(db_path, 'time_series_19-covid-Confirmed.csv'), index_col=indexes)
df_recovered = pd.read_csv(os.path.join(db_path, 'time_series_19-covid-Recovered.csv'), index_col=indexes)
df_deaths = pd.read_csv(os.path.join(db_path, 'time_series_19-covid-Deaths.csv'), index_col=indexes)
```

```python
df = pd.DataFrame({
    'Total Confirmed': df_confirmed.stack(),
    'Total Deaths': df_deaths.stack(),
    'Total Recovered': df_recovered.stack()
})

df.index.rename('Date', level=4, inplace=True)

df.reset_index()
```

```python
df.to_csv('data.csv')
```
